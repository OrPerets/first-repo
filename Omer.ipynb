```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

A = 'Wetland Out'
B = 'Agam'
C = 'Maayan'
D = 'Nahal Hadar'
```

```python
data = pd.read_excel('data.xlsx')
data.set_index("Test", inplace=True)
data[[A,B,C,D, 'Wetland In']] = data[[A,B,C,D, 'Wetland In']].astype(float)
```

```python
data.dropna().groupby('Year')[A].mean().plot(kind="pie", autopct='%.2f%%', figsize=(8,6), fontsize=16)
```

```python
data["Test"] = data.index
data.index = range(len(data))
```

```python
data
```

```python
years = data.groupby(['Test'])
for name, group in years:
    group.groupby('Year')[[A,B,C,D]].mean().plot(figsize=(10,10), fontsize=18, title=name)
    plt.legend(prop={'size': 11})
```

```python
years = data.groupby(['Year'])
for name, group in years:
    group = group[(group["Test"] != 'Ecoli') & (group["Test"] != 'coli F')]
    group.groupby('Test')[[A,B,C,D]].mean().plot(figsize=(10,10), fontsize=18, title=name, rot=90)
    plt.legend(prop={'size': 11})
```
